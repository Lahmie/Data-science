```python
%matplotlib inline
import pandas as pd
pd.plotting.register_matplotlib_converters()
from pandas import DataFrame
import matplotlib as mpl
import seaborn as sns 
import matplotlib.pyplot as plt
```

```python
production = pd.read_csv("Africa Food Production (2004 - 2013).csv")
```

```python
supply = pd.read_csv("Africa Food Supply (2004 - 2013).csv")
```

```python
import warnings
warnings.filterwarnings('ignore')
```

```python
plt.figure(figsize=(16,6))
sns.relplot(x='Value',y='Year',hue='Country',kind='line', data=production);
sns.relplot(x='Value',y='Year',hue='Item',kind='line', data=production);
```

```python
g = sns.FacetGrid(production, col='Country',hue='Country')
g = g.map(sns.barplot,'Value','Year')
g = g.map(sns.pointplot, 'Value', 'Year')
g
```
